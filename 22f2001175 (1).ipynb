#### Since there is no feature in which there is high value of correlation between them so we cannot drop any feature

```python
def detect_outliers_iqr(df):
    results = []
    
    for col in df.select_dtypes(include=np.number):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        results.append([col, len(outliers)])
    
    return pd.DataFrame(results, columns=["Feature", "Outlier_Count"]).sort_values(
        "Outlier_Count", ascending=False
    )
outliers_df = detect_outliers_iqr(data)
print(outliers_df)
```

```python
import re
import pandas as pd

def clean_text(text):
    if pd.isna(text):        
        return ""
    text = str(text)         
    text = text.lower()
    text = re.sub(r'http\S+', '', text)        
    text = re.sub(r'[^a-z0-9\s]', '', text)   
    text = re.sub(r'\s+', ' ', text).strip()  
    return text
data['clean_comment'] = data['comment'].apply(clean_text)
test['clean_comment'] = test['comment'].apply(clean_text)
```

# **Pipeline**

```python
text_feature = ["clean_comment"]
#categorical_features = ["religion", "race", "gender"]  
numeric_features = [
    "engagement_score", "disability",
    "emoticon_1", "emoticon_2", "emoticon_3",
    "if_1", "if_2", "post_id", "day_of_week", "hour"
]
```

```python
cat_pipe=Pipeline([
    ("encoder", OneHotEncoder(handle_unknown = "ignore"))
])
cat_pipe
```

```python
num_pipe = Pipeline([  
    ("scaler", RobustScaler())
])
num_pipe
#robust scaler and drop category columns
```

```python

text_pipe = Pipeline([
    ("flatten", FunctionTransformer(lambda x: x.squeeze(), validate=False)),
    ("tfidf", TfidfVectorizer(
        max_features=50000,
        ngram_range=(1,3),
        stop_words="english"
    ))
])
```

```python

preprocessor = ColumnTransformer([
    ("text", text_pipe, text_feature),
    ("num", num_pipe, numeric_features)
])
```

```python
model_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1500, class_weight='balanced'))
])

```

```python
model_pipeline.fit(X_train, y_train)
```

```python
y_pred = model_pipeline.predict(X_test)
```

```python
print(y_pred)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from lightgbm import LGBMClassifier

model_pipeline_2 = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LGBMClassifier(
        boosting_type='gbdt',  
        n_estimators=1000,     
        learning_rate=0.05,    
        max_depth=-1,          
        class_weight='balanced',
        random_state=42
    ))
])
```

```python
model_pipeline_2.fit(X_train, y_train)
```

```python
y_pred2 = model_pipeline_2.predict(X_test)
```

```python
print(y_pred2)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred2))
```
